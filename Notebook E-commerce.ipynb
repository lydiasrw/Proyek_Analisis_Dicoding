## Cleaning order_df

```python
#Change data type time in order_df

order_df.info()
```

```python
order_df.head(10)
```

# Visualization dan Exploratory

## Bagaimana penjualan dan revenue selama tiga tahun terakhir (2016-2018)?

```python
year_order_df = commerce_df.resample(rule="Y", on="order_purchase_timestamp").agg({
    "order_id" : "nunique",
    "price" : "sum"
})

year_order_df.index = year_order_df.index.strftime("%Y") #Mengubah data datetime menjadi tahun
year_order_df = year_order_df.reset_index()
year_order_df.rename(columns={
    "order_id" : "count_id",
    "price" : "revenue"
}, inplace=True)


year_order_df.head()
```

```python
#Membuat line chart untuk melihat total penjualan 

plt.figure(figsize=(10,5))
plt.plot(year_order_df["order_purchase_timestamp"], year_order_df["count_id"], marker="o", linewidth=1, color="#D988B9")
plt.title("Total Order per Year", loc="center", fontsize=12)
plt.xticks(fontsize=8)
plt.yticks(fontsize=10)

plt.show()
```

```python
#Membuat line chart untuk melihat revenue penjualan selama tiga tahun (2016-2018)

plt.figure(figsize=(10,5))
plt.plot(year_order_df["order_purchase_timestamp"], year_order_df["revenue"], marker="o", linewidth=1, color="#D988B9")
plt.title("Revenue Order per Year", loc="center", fontsize=12)
plt.xticks(fontsize=8)
plt.yticks(fontsize=10)

plt.show()
```

## Produk apa yang paling banyak terjual selama tahun 2016-2018?

```python
#Produk yang paling banyak terjual

sum_product_df = commerce_df.groupby(by="product_category_name").agg({
    "freight_value" : "sum",
    "price" : "sum"
}).sort_values(by="freight_value", ascending=False).reset_index()

sum_product_df.rename(columns={
    "freight_value" : "count",
    "price" : "income"
}, inplace=True)

sum_product_df.head(5)
```

```python
#Visualisasi produk terjual berdasarkan jumlah produk

plt.figure(figsize=(10,5))
color=["#776B5D", "#EBE3D5", "#EBE3D5", "#EBE3D5", "#EBE3D5"]
sn.barplot(x="count", y="product_category_name", data=sum_product_df.sort_values(by="count", ascending=False).head(5), palette=color)

plt.title("The Best Product", loc="center", fontsize=10)
plt.xlabel("Count")
plt.ylabel("Product")

plt.show()
```

```python
#Visualisasi produk terjual berdasarkan jumlah income

plt.figure(figsize=(10,5))
color=["#776B5D", "#EBE3D5", "#EBE3D5", "#EBE3D5", "#EBE3D5"]
sn.barplot(x="income", y="product_category_name", data=sum_product_df.sort_values(by="income", ascending=False).head(5), palette=color)

plt.title("The Best Product", loc="center", fontsize=10)
plt.xlabel("Income")
plt.ylabel("Product")

plt.show()
```

## Kota apa yang mendapatkan income terbesar selama tahun 2016-2018?

```python
sum_seller_df = commerce_df.groupby(by="seller_city").price.sum().sort_values(ascending=False).reset_index()
sum_seller_df.head()
```

```python
#Barchart untuk "The best seller in the city"

plt.figure(figsize=(10,5))
color=["#94A684", "#E4E4D0", "#E4E4D0", "#E4E4D0", "#E4E4D0"]
sn.barplot(x="seller_city", y="price", data=sum_seller_df.head(5), palette=color)

plt.title("The Best Selling in the City", loc="center", fontsize=12)
plt.xlabel("City")
plt.ylabel(None)

plt.show()
```

## Kota mana dengan pelanggan terbanyak? 

```python
sum_customer_df = commerce_df.groupby(by="customer_state").customer_id.nunique().sort_values(ascending=False).reset_index()
sum_customer_df.head()
```

```python
#Demografi pelanggan berdasarkan state

plt.figure(figsize=(10,5))
color=["#E9B824", "#F4BF96", "#F4BF96", "#F4BF96", "#F4BF96"]
sn.barplot(x="customer_state", y="customer_id", data=sum_customer_df.head(5), palette=color)

plt.title("Demography of Customers", loc="center", fontsize=12)
plt.xlabel("City")
plt.ylabel(None)

plt.show()
```
